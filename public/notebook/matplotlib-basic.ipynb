```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]
```

```python
plt.scatter(x_values,y_values) # 绘制散点图
```

```python
plt.scatter(x_values,y_values, marker='x', c='red')  # marker 设置形状，c 设置颜色
```

```python
plt.plot(x_values,y_values)  # 绘制折线图
```

```python
def plot_segment(p1,p2):
    x1,y1 = p1
    x2,y2 = p2
    plt.plot([x1,x2],[y1,y2],marker='o')
    
point1 = (0,4)
point2 = (6,1)

plot_segment(point1,point2)
```

```python
x_values = np.arange(0,10,0.01)
y_values = np.sin(x_values)
plt.plot(x_values,y_values)
```

```python
# 默认绘制，不是按比例绘制，通过ylim，xlim设置 x，y轴的比例

plt.ylim(0,5) 
plt.xlim(0,5)

point1 = (0,3)
point2 = (2,1)

plot_segment(point1,point2)
```

```python
x_values = np.arange(0,10,0.01)
y_values = np.sin(x_values)

plt.plot(x_values,y_values)
plt.title("Graph of sin(x) vs x", fontsize=16)  # 设置标题
plt.xlabel("This is the x value", fontsize=16)  # 设置 x 轴标签
plt.ylabel("The value of sin(x)", fontsize=16)  # 设置 y 轴标签
```
